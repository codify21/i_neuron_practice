```python
#importing basics libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset =pd.read_excel('Data_Train.xlsx')
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset['Date_of_Journey'].str.split('/').str[0]
```

```python
dataset['Date']=dataset['Date_of_Journey'].str.split('/').str[0]
dataset['Month']=dataset['Date_of_Journey'].str.split('/').str[1]
dataset['Year']=dataset['Date_of_Journey'].str.split('/').str[2]
dataset.head(3)
```

```python
dataset["Date"]=dataset['Date_of_Journey'].apply(lambda x:x.split("/")[0])
dataset["Month"]=dataset['Date_of_Journey'].apply(lambda x:x.split("/")[1])
dataset["Year"]=dataset['Date_of_Journey'].apply(lambda x:x.split("/")[2])
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset["Date"]=dataset["Date"].astype(int)
dataset["Month"]=dataset["Month"].astype(int)
dataset["Year"]=dataset["Year"].astype(int)
```

```python
dataset.head(2)
```

```python
dataset.info()
```

```python
dataset.drop('Date_of_Journey',inplace =True,axis=1 )#inplace will modify in the dataset iteslf  
#inplace = True changes the default behaviour such that the operation on the dataframe doesn't return anything,
#it instead 'modifies the underlying data'
```

```python
dataset.head(2)
```

```python
dataset['Arrival_Time'].str.split(' ').str[0]
```

```python
dataset['Arrival_Time']=dataset['Arrival_Time'].apply(lambda x : x.split(' ')[0])
dataset.head(2)
```

```python
dataset['Arrival_hour']=dataset['Arrival_Time'].str.split(':').str[0]
dataset['Arrival_min']=dataset['Arrival_Time'].str.split(':').str[1]
dataset.head(2)
```

```python
dataset.info()
```

```python
dataset["Arrival_hour"]=dataset["Arrival_hour"].astype(int)
dataset["Arrival_min"]=dataset["Arrival_min"].astype(int)
dataset.info()
```

```python
dataset.drop('Arrival_Time',inplace =True,axis=1 )
dataset.head(2)
```

```python
dataset['Dept_hour']=dataset['Dep_Time'].str.split(':').str[0]
dataset['Dept_min']=dataset['Dep_Time'].str.split(':').str[1]
dataset['Dept_hour']=dataset['Dept_hour'].astype(int)
dataset['Dept_min']=dataset['Dept_min'].astype(int)
dataset.drop('Dep_Time',axis=1,inplace=True)
```

```python
dataset.head(2)
```

```python
dataset['Total_Stops'].unique()
```

```python
dataset['Total_Stops']=dataset['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
#nan=1 (mode = 1)
```

```python
dataset.head(2)
```

```python
dataset['Total_Stops'].unique()
```

```python
dataset[dataset['Total_Stops'].isnull()]
```

```python
dataset.drop('Route',axis=1,inplace=True)
```

```python
dataset.head(2)
```

```python
dataset['Source'].unique()
```

```python
dataset['Destination'].unique()
```

```python
dataset['Additional_Info'].unique()
```

```python
dataset=pd.get_dummies(dataset,columns=['Airline','Source','Destination','Additional_Info'],drop_first=True)
# used inplace of one hot encoding we are using pandas get_dummies function
```
